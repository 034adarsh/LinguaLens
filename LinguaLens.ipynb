```python
from transformers import pipeline

# Load the Arabic-to-English translation pipeline
translator = pipeline("translation_ar_to_en", model="Helsinki-NLP/opus-mt-ar-en")

# Test the model with a simple string
text = "مرحبا بك في فريق EY"
translated_text = translator(text, max_length=400)[0]['translation_text']
print("Translated Text:", translated_text)
```

```python
import pandas as pd

def translate_excel(file_path):
    # Read the Excel file
    df = pd.read_excel("gl_data_arabic.xlsx")

    # Translate column headers
    translated_columns = {col: translator(col, max_length=400)[0]['translation_text'] for col in df.columns}

    # Rename columns
    df.rename(columns=translated_columns, inplace=True)

    return df
```

```python
# Call the function with the file path
translated_df = translate_excel("gl_data_arabic.xlsx")
translated_df.to_excel("translated_gl_data.xlsx", index=False)
print("Translation completed and saved as 'translated_gl_data.xlsx'")
```
